The purpose of the study is to see if inflation has an impact on the prices of the various sectors of real estate [ie. Residential/Commercial/Industrial]

We decided to compare inflation rate to industrial [warehouse], commercial [office] and residential [EC] properties from 2017 September - 2022 September

Inflation is defined as buying power of each dollar being reduced due to an increase in price over a given period of time.

Real Estate is affected by limited availability, housing construction cost and use as income-generating assets. Utilized as an inflation hedge, the intention is to advise which sector would be most viable as an asset.




```python
'''
IMPORTS
'''
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import requests
import datetime as dt
import statsmodels.formula.api as smf
import geopandas
```

## Inflation data

```python
#Webscraping Singapore's inflation data from rateinflation.com
inflation_df = pd.read_html(requests.get('https://www.rateinflation.com/inflation-rate/singapore-historical-inflation-rate/').text)

#Select inflation data
inflation_df = inflation_df[0]

#Get data from 2017-2022 in reverse order
inflation_df = inflation_df.loc[5::-1]

#Create values for x-axis for months
months = pd.period_range("2017-01", "2022-09" , freq='M')

#Append inflation data to a list
lst = []
annual = []
for i in inflation_df.to_numpy():
    #Getting only monthly inflation rate from data table
    lst.extend(i[1:-1])
    annual.append(i[-1])


#Removing last 3 months in 2022 because no data yet
lst = lst[:-3]

#Removing % from the inflation data and changing data type to float
lst = [float(i[:-1]) for i in lst]
```

```python
#Inflation Data
inflation_data = {}
for t, rate in zip(months,lst):
    inflation_data[str(t)] = rate
```

```python
#Plotting the inflation rate in Singapore
plt.plot(months.to_timestamp(),lst)
plt.ylabel("Inflation Rate (%)")
plt.title("Inflation rate (Monthly) in Singapore")
plt.grid((0.9,0.9,0.9))
```

Based on the inflation rate gathered from the period of 2017 to 2022 monthly, the recent years from Q1 of 2020 has shown a drastic rise in inflation to upwards of ~7.5%.


#### Inflation today is caused by 2 reasons:

1. Excess money in the system when monetary authorities increased the supply of money at the beginning stages of the pandemic to boost the economy.
2. Supply of goods is unable to meet the strong demand. During the pandemic, demand for goods increased drastically. As the borders reopen, the supply for goods is unable to meet the increasing demand for goods and services. Coupled with supply chain limitations, lockdown in China and the increased energy prices (Ukraine-Russia war), these factors drive up costs. 

## Market Data
There are 3 market datasets representing 3 different sectors:
1. Residential: represented by executive condo
2. Commericial: represented by offices
3. Industrial: represented by warehouses

### Key Variables

`ALL_DFS`\
Type: `Dict[str, pd.DataFrame]`
- this is a dictionary containing all the dataframes, values and dataframes inside will __NOT change after the cell it is declared in__ . It's values consists of the following
    1. residential_df
    2. commercial_df
    3. industrial_df


```python
'''
Directories to csv datasets

Variables declared here are all str, they are NOT the actual dataframe
'''
DATA_DIR:str = "Data"

RESIDENTIAL_CSV:str = f"{DATA_DIR}/ExecutiveCondoTransaction.csv"
COMMERCIAL_CSV:str = F"{DATA_DIR}/OfficeTransaction.csv"
INDUSTRIAL_CSV:str = F"{DATA_DIR}/WarehouseTransaction.csv"
```

```python
'''
Initialising the actual dataframes and ALL_DFS
'''
# Reading the csv -> converting into pd.DataFrame -> storing into ALL_DFS
ALL_DFS = {
    "residential": pd.read_csv(RESIDENTIAL_CSV),
    "commercial": pd.read_csv(COMMERCIAL_CSV),
    "industrial": pd.read_csv(INDUSTRIAL_CSV)
}

residential_df: pd.DataFrame = ALL_DFS["residential"]
commercial: pd.DataFrame = ALL_DFS["commercial"]
industrial_df: pd.DataFrame = ALL_DFS["industrial"]

# Data cleaning
for df_name in ALL_DFS: #for each df
    df = ALL_DFS[df_name]
    df = df.dropna() # drop all na values
    df["Sale Date"] = pd.DatetimeIndex(df["Sale Date"]).year # parsing the dates

    df["Unit Price ($ PSF)"] = df["Unit Price ($ PSF)"].replace(",", "", regex=True) # removing the commas in prices
    df["Unit Price ($ PSF)"] = pd.to_numeric(df["Unit Price ($ PSF)"], errors = 'coerce') # converting the price values from strings to numeric
    ALL_DFS[df_name] = df
```

## Mean unit prices for different sectors

This section studies the mean unit prices for each sectors (residential, commercial and industrial) between 2017 and 2022

```python
'''
Initialising dataframes for this section

dataframes are copied from ALL_DFS and then processed in this new set of dataframes so that the global ALL_DFS remains unchanged
'''
all_dfs_groupby_sale_date = {} # dataframes to use for this section
for df_name in ALL_DFS:
    all_dfs_groupby_sale_date[df_name] = ALL_DFS[df_name].groupby("Sale Date").mean(numeric_only = True) # group by sale date so that mean prices for each date can be calculated
```

```python
'''
Plotting
'''
plt.figure(figsize=(10,5))
for i, df_name in enumerate(all_dfs_groupby_sale_date): # for each sector,
    y = all_dfs_groupby_sale_date[df_name]["Unit Price ($ PSF)"] # retrieve mean unit price for each sale date
    plt.plot(y, label=df_name)
    plt.legend()
    
plt.title('Mean Unit Prices For Different Sectors')
plt.xlabel('Year')
plt.ylabel('Mean Unit Price ($ PSF)')
plt.grid()
plt.show()
```

Shown on the mean unit prices for the 3 different sectors, there's a correlation where the increase in inflation has caused the price of each mean unit in each industry to increase as well from 2020 onwards.

## Mean unit prices for each planning regions

Taking a closer study at the 3 different sectors, we split the data into their respective planning regions.

### Planning regions
1. Central Region
2. East Region
3. North East Region
4. North Region
5. West Region

```python
# the planning regions to split into
planning_regions = [
    "Central Region",
    "East Region",
    "North East Region",
    "North Region",
    "West Region"
]

# number of subplots
n_rows = 3 # number of rows
n_columns = 2   # number of columns

fig, axs = plt.subplots(n_rows, n_columns, figsize = (10,10))
fig.suptitle('Mean prices of various sectors')

for subplot_index, planning_region in enumerate(planning_regions): # for each planning region
    row = subplot_index//n_columns  # the row number of the subplot
    column = subplot_index % n_columns # column number of the subplot
    
    '''************************************************************
    Start of subplot (each planning region is 1 subplot)
    ************************************************************'''
    for i, df_name in enumerate(ALL_DFS): # for each sector
        sector_df = ALL_DFS[df_name]
        df = sector_df.loc[sector_df["Planning Region"] == planning_region] # retrieve the current planning region determined in the outer for loop
        df = df.groupby("Sale Date").mean(numeric_only = True) # group by sale date so we can retrieve aggregate data for each date
        y = df["Unit Price ($ PSF)"] # retrieve the mean unit prices
        
        # plot
        subplot = axs[row, column]
        subplot.plot(y, label=df_name)
        subplot.title.set_text(planning_region)
        subplot.legend()
    '''************************************************************
    End of subplot
    ************************************************************'''

'''
Deleting extra empty subplot canvas

Since we declared number of rows = 3 and number of columns = 2, there is a total of 6 subplot canvas
but we only have 5 setcors, i.e. 5 subplots

Hence, there is 1 additional empty canvas which have to be deleted
'''
if num_empty_plots := (n_rows * n_columns) - len(planning_regions): # calculating the number of extra subplots
    for i in range(num_empty_plots): # for each extra subplots
        fig.delaxes(axs[-1,-1-i]) # delete the subplot

plt.tight_layout()
plt.show()
```

Viewing each sector through the different planning regions of Singapore, the residential sector is the most resilient with increases in each region. Notably, while each region has a slow incline, the north region experiences a steep price increase, likely due to the resumption of construction and subsequent opening of various facilities such as Bukit Canberra which increases the value of the residential property.

The industrial and commercial sectors all experience a decline, with an increase at the tail end. While the industrial sector is rigid with barely any price volatility, the commercial faces the most changes regardless of region. 

# Regression

```python
regression_residential_df = residential_df[["Unit Price ($ PSF)","Sale Date"]]
regression_residential_df = regression_residential_df.rename(columns={"Unit Price ($ PSF)": "Price", "Sale Date":"Date"})

regression_residential_df["Price"] = regression_residential_df["Price"].replace(",", "", regex=True)
regression_residential_df['Price'] = pd.to_numeric(regression_residential_df['Price'], errors = 'coerce')

regression_residential_df['Date'] = pd.to_datetime(regression_residential_df['Date'])
regression_residential_df['month_year'] = regression_residential_df['Date'].dt.to_period('M')
regression_residential_df['month_year'] = regression_residential_df["month_year"].astype('str')
regression_residential_df['rate'] = regression_residential_df['month_year'].apply(lambda x: inflation_data[x])
```

```python
regression_industrial_df = industrial_df[["Unit Price ($ PSF)","Sale Date"]]
regression_industrial_df = regression_industrial_df.rename(columns={"Unit Price ($ PSF)": "Price", "Sale Date": "Date"})
regression_industrial_df["Price"] = regression_industrial_df["Price"].replace(",", "", regex=True)
regression_industrial_df['Price'] = pd.to_numeric(regression_industrial_df['Price'], errors ='coerce')
regression_industrial_df['Date'] = pd.to_datetime(regression_industrial_df['Date'])
regression_industrial_df['month_year'] = regression_industrial_df['Date'].dt.to_period('M')
regression_industrial_df['month_year'] = regression_industrial_df["month_year"].astype('str')
regression_industrial_df['rate'] = regression_industrial_df['month_year'].apply(lambda x: inflation_data[x])
```

```python
regression_commercial_df = commercial_df[["Unit Price ($ PSF)","Sale Date"]]
regression_commercial_df = regression_commercial_df.rename(columns={"Unit Price ($ PSF)": "Price", "Sale Date": "Date"})
regression_commercial_df["Price"] = regression_commercial_df["Price"].replace(",", "", regex=True)
regression_commercial_df['Price'] = pd.to_numeric(regression_commercial_df['Price'], errors ='coerce')
regression_commercial_df['Date'] = pd.to_datetime(regression_commercial_df['Date'])
regression_commercial_df['month_year'] = regression_commercial_df['Date'].dt.to_period('M')
regression_commercial_df['month_year'] = regression_commercial_df["month_year"].astype('str')
regression_commercial_df['rate'] = regression_commercial_df['month_year'].apply(lambda x: inflation_data[x])
```

```python
# Linear relationship (Linear trendline)
reg1 = smf.ols("Price ~ rate ", data = regression_residential_df).fit()
regression_residential_df['fit_price'] = reg1.params[0] + reg1.params[1]*regression_residential_df['rate']

plt.figure(figsize = (8,4))
plt.subplot(1,1,1)
plt.scatter(regression_residential_df['rate'], regression_residential_df['Price'],s=4, label = 'Data')
plt.plot(regression_residential_df['rate'],regression_residential_df['fit_price'],color=(0,0,0), label = 'Linear fit')
plt.title('Effect of inflation rate on unit price (PSF) of residential properties')
plt.xlabel('Inflation Rate (Monthly)')
plt.ylabel('Unit Price ($ PSF)')
plt.legend()
plt.grid(color=(.9,.9,.9))
plt.show()
```

```python
reg2 = smf.ols("Price ~ rate ", data = regression_industrial_df).fit()
regression_industrial_df['fit_price'] = reg2.params[0] + reg2.params[1]*regression_industrial_df['rate']

plt.figure(figsize = (8,4))
plt.subplot(1,1,1)
plt.scatter(regression_industrial_df['rate'], regression_industrial_df['Price'],s=4, label = 'Data')
plt.plot(regression_industrial_df['rate'],regression_industrial_df['fit_price'],color=(0,0,0), label = 'Linear fit')
plt.title('Effect of inflation rate on unit price (PSF) of industrial properties')
plt.xlabel('Inflation Rate (Monthly)')
plt.ylabel('Unit Price ($ PSF)')
plt.legend()
plt.grid(color=(.9,.9,.9))
plt.show()
```

```python
reg3 = smf.ols("Price ~ rate ", data = regression_commercial_df).fit()
regression_commercial_df['fit_price'] = reg3.params[0] + reg3.params[1]*regression_commercial_df['rate']

plt.figure(figsize = (8,4))
plt.ylim([0,4000])
plt.subplot(1,1,1)
plt.scatter(regression_commercial_df['rate'], regression_commercial_df['Price'],s=4, label = 'Data')
plt.plot(regression_commercial_df['rate'],regression_commercial_df['fit_price'],color=(0,0,0), label = 'Linear fit')
plt.title('Effect of inflation rate on unit price (PSF) of commercial properties')
plt.xlabel('Inflation Rate (Monthly)')
plt.ylabel('Unit Price ($ PSF)')
plt.legend()
plt.grid(color=(.9,.9,.9))
plt.show()
```

Based on the regression plot that shows a positive relation between the unit price and inflation rate, regardless of the types of property, with the residential properties having the steepest increase in unit price.

When supply is limited, inflation drives prices up, regardless of the goods. This is true for the property market as well. Quoting an analogy from Investopedia, if there are only 5 homes in the market, with \$10 of currency in the market, then each home is worth \$2 each. If there is more money in the system, for e.g. \$20 of currency, consumers can now afford to pay more for the same home and each home is worth \$4 now. A rising inflation pushes prices up only if there is only so much supply of goods in the market. 

This correlation can be observed from past data. Thus, because of this correlation, real estate is a preferred asset class. Owning a piece of real estate in an inflationary environment preserves the capital because the real value of the property adjusts for inflation.

In summary, the real estate market is right in the perfect storm of high inflation, high-interest rate, rising local and foreign demand and dwindling supply of new homes. This imbalance of supply and demand is mainly why we are witnessing the accelerating rate in property prices.

```python
# Extracting Unit Price ($ PSF) > 1000 data and Postal Code data from residential dataframe
map_df = residential_df[["Unit Price ($ PSF)", "Postal Code"]]
map_df["Price"] = map_df["Unit Price ($ PSF)"].replace(",", "", regex=True)
map_df['Price'] = pd.to_numeric(map_df['Price'], errors = 'coerce')
map_df = map_df.loc[map_df['Price']>1000]
map_df = map_df.reset_index(drop=True)
```

```python
import time
import json
# import simplejson as json
```

```python
# Running postal codes through OneMap API to get latitude and longitude of buildings
# Takes long to run (Around 250 entries per min), so highly recommended to save to another file when done. 
# Saved to mapdata.csv in this case
saved = True

if saved:
    map_df = pd.read_csv("Data/mapdata.csv")
else:
    for i in range(len(map_df)):
        if len(str(map_df['Postal Code'][i]))==5:
            map_df['Postal Code'][i]='0'+map_df['Postal Code'][i]
            
    map_df['OnemapLongitude'] = pd.Series(0, index=map_df.index)
    map_df['OnemapLatitude'] = pd.Series(0, index=map_df.index)       
    map_df['Postal Code'] = map_df['Postal Code'].astype(str)

    start=time.time()
    for i in range(len(map_df)):
        req=requests.get('https://developers.onemap.sg/commonapi/search?searchVal='+map_df['Postal Code'][i]+'&returnGeom=Y&getAddrDetails=Y&pageNum=1')
        jdata = json.loads(req.text)

        if jdata['found']>=1:
            map_df['OnemapLongitude'][i]=jdata['results'][0]['LONGITUDE']
            map_df['OnemapLatitude'][i]=jdata['results'][0]['LATITUDE']
        else:
            map_df['OnemapLongitude'][i]=float('nan')
            map_df['OnemapLatitude'][i]=float('nan')
        print(f"{i}/{len(map_df)}")
        
    print('Time Taken:',time.time()-start)
    map_df.to_csv("Data/mapdata.csv")
```

```python
# Read latitude and longitude data from generated csv and convert the format into epsg:4326 latitude/longitude coordinate system
df_codes = geopandas.read_file('Data/mapdata.csv')
df = geopandas.GeoDataFrame(df_codes, geometry=geopandas.points_from_xy(df_codes["OnemapLongitude"].astype('float32'), df_codes["OnemapLatitude"].astype('float32')) )
df.crs = {'init': 'epsg:4326'}

# Read the Singapore map
gf0 = geopandas.read_file('zip://data/stanford-pg798kr1205-shapefile.zip')
```

```python
# Plotting the map figure
fig, axis = plt.subplots(1,1,figsize=(10,10))
gf0.plot(ax=axis, color=(.8,.8,.8))
df.plot(ax = axis,marker='d',markersize=5)
plt.xticks([])
plt.yticks([])
plt.axis('off')
plt.title('Residential Properties with Unit Price ($PSF) > SGD1000',fontsize=13)
plt.show()
```
