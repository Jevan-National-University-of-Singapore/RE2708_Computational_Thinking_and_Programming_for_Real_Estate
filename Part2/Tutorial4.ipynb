```python
import pandas as pd
import re

DATA_FILE = "Data/hdb-property-information.csv"

df = pd.read_csv(DATA_FILE)

df.dropna(inplace=True) # data cleaning

display(df.head())
```

```python
count_df = df.groupby("street").count()
count_df.head()
```

```python
import matplotlib.pyplot as plt


plt.figure(figsize= (10,10))
plt.bar(df.street.unique(), count_df["residential"])
plt.show()
```
